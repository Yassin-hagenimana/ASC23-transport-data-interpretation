```python
# YOUR CODE HERE
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='co2', bins=30, kde=True, color='grey')
plt.title('Distribution of CO2 Emissions')
plt.xlabel('CO2 Emissions (g/mile)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# YOUR CODE HERE
#Interpretations
# The histogram of CO2 emissions shows that most vehicles emit between 300 and 450 grams/mile of carbon dioxide.
#The distribution is slightly right-skewed, meaning there are fewer vehicles with very high CO2 emissions (above 500 g/mile).
#This suggests that the majority of vehicles in the dataset are relatively moderate emitters, but there are some high-emission outliers (likely performance or large-engine vehicles).
```

```python
# YOUR CODE HERE
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='comb', bins=25, kde=True, color='orange')
plt.title('Distribution of Combined Fuel Efficiency')
plt.xlabel('Combined MPG')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#Interpretation:
#The combined fuel efficiency (MPG) histogram peaks between 20 and 30 MPG, meaning most vehicles fall within this efficiency range.
#The curve is left-skewed, indicating fewer fuel-inefficient vehicles (under 15 MPG).
#This distribution suggests that the dataset mostly includes vehicles with average to above-average fuel efficiency, with fewer extremely low-efficiency models like sports cars or large trucks.

```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='displ', bins=20, kde=True, color='skyblue')
plt.title('Distribution of Engine Displacement')
plt.xlabel('Displacement (Liters)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#Interpretation:
#The engine displacement histogram reveals that engines around 2.0 to 3.5 liters are the most common.
#There are fewer vehicles with very small engines (under 1.5L) and very large engines (above 4.5L).
#This suggests that most vehicles in the dataset use mid-size engines, which balance power and fuel efficiency. The presence of some high-displacement engines indicates performance or utility-focused vehicles.
```

```python
df_numeric = df.select_dtypes(include=[np.number])

corr_matrix = df_numeric.corr()

cmap = sns.diverging_palette(20, 200, as_cmap=True)

sns.heatmap(corr_matrix,
            cmap=cmap,
            vmax=1.0,
            vmin=-1.0,
            center=0,
            square=True,
)

plt.tight_layout()
plt.show()
```

# Our intepretation for this chart

## What the Data Shows:

1. **Engine Size vs. Fuel Efficiency**:
Cars with bigger engines (more cylinders and higher displacement) generally get worse gas mileage and produce more CO2. This makes sense - larger engines need more fuel to run.

2. **City and Highway MPG**:
When a car gets good city mileage, it usually gets good highway mileage too. This shows that efficient cars tend to be efficient everywhere, not just in one type of driving.

3. **Emissions and Ratings**:
Cars with lower CO2 emissions always have better environmental scores (ghgScore). This direct relationship shows that the rating system correctly rewards cleaner vehicles.
